```python
import pandas as pd
import numpy as np

# use pandas read_csv function to read the csv file and pandas DataFrame method to convert file into the data frame

df = pd.DataFrame(pd.read_csv('/content/titanic.csv'))
df.head()

# use shape is returning no. of rows and no. of column

df.shape

# isnull() is used to calculate no. of null value in each column

df.isnull().sum()

# seperating the column which have more than 35% vlaue missing

drop_col = df.isnull().sum()[df.isnull().sum()>(35/100 * df.shape[0])]
drop_col

#we drop the cabin column because it not much usefull

drop_col.index
df.drop(drop_col.index, axis=1, inplace=True)
df.isnull().sum()

# we fill all the null value with mean

df.fillna(df.mean(), inplace=True)
df.isnull().sum()

# to know about each column

df['Embarked'].describe()

# for Embarked fill the null value with most frequent value in column

df['Embarked'].fillna('S', inplace=True)

# for correlation

df.corr()

# we create new column using two existing column

df['FamilySize'] = df['SibSp']+df['Parch']
df.drop(['SibSp','Parch'],axis=1,inplace=True)
df.corr()

# check person is alon or not

df['Alone'] = [0 if df['FamilySize'][i]>0 else 1 for i in df.index]
df.head()
df.groupby(['Alone'])['Survived'].mean()

df[['Alone','Fare']].corr()

# 1 for female ,0 for male

df['Sex']=[0 if df['Sex'][i]=='male' else 1 for i in df.index]
df.groupby(['Sex'])['Survived'].mean()

df.groupby(['Embarked'])['Survived'].mean()

#CONCLUSION

#Female passengers were prioritized over man
#people with high class or rich people have higher survivl rate than others
#passenger travelling with their family have higher survivl rate
#passenger who borded the ship at cherbourg survive more in prapotion than other











```
